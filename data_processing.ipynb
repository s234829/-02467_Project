```python
import json
import pandas as pd
from tqdm import tqdm
```

Loading in the dataset <br>
From the structure data/jsonl file (so not from the gz files)

```python
def load_reviews(file):
    data = []
    with open(file, 'r', encoding='utf-8') as fp:
        for line in tqdm(fp, desc='Loading reviews'):
            try:
                dp = json.loads(line.strip())
                
                rating = float(dp.get('rating', 0))
                title = dp.get('title', '')
                text = dp.get('text', '')
                images = dp.get('images', [])
                asin = dp.get('asin', '')
                parent_asin = dp.get('parent_asin', '')
                user_id = dp.get('user_id', '')
                timestamp = int(dp.get('timestamp', 0))
                helpful_vote = int(dp.get('helpful_vote', 0))
                verified = bool(dp.get('verified_purchase', False))

                data.append((
                    rating, title, text, images, asin, parent_asin,
                    user_id, timestamp, helpful_vote, verified
                ))
            except ValueError as e:
                print(f"Skipping line due to error: {e}\n{line}")
    return data
# pd.DataFrame(data, columns=[
    "rating", "title", "text", "images", "asin", "parent_asin",
    "user_id", "timestamp", "helpful_vote", "verified_purchase"#])
```

```python
file_review = "data/Movies_and_TV.jsonl"
data_review = load_reviews(file_review)
df_review = pd.DataFrame(data_review, columns=[
    "rating", "title", "text", "images", "asin", "parent_asin",
    "user_id", "timestamp", "helpful_vote", "verified_purchase"])
```

```python
def load_metadata(file):
    data = []
    with open(file, 'r', encoding='utf-8') as fp:
        for line in tqdm(fp, desc='Loading metadata'):
            try:
                dp = json.loads(line.strip())

                main_category = dp.get('main_category', '')
                title = dp.get('title', '')
                subtitle = dp.get('subtitle', '')
                average_rating = dp.get("average_rating", 0)
                rating_number = dp.get("rating_number", 0)
                features = dp.get('features', [])
                description = dp.get('description', [])
                price = dp.get('price', "")
                parent_asin = dp.get('parent_asin', '')
                bought_together = dp.get('bought_together')
                images = dp.get("images", [])

                # # Extract first 720w image URL if present
                # image_url = ''
                # resolutions = ['1920w', '1440w', '1080w', '720w', '480w', '360w']

                # for img in dp.get('images', []):
                #     if isinstance(img, dict):
                #         for res in resolutions:
                #             if res in img:
                #                 image_url = img[res]
                #                 break
                #     if image_url:  # Found one
                #         break
                
                # Flatten details
                details = dp.get('details', {})
                # content_advisory = details.get('Content advisory', [])
                # audio_languages = details.get('Audio languages', [])
                # subtitles = details.get('Subtitles', [])
                # directors = details.get('Directors', [])

                data.append((
                    main_category, title, subtitle, average_rating, rating_number,
                    features, description, price, images,
                    details,parent_asin, bought_together
                ))

            except json.JSONDecodeError as e:
                print(f"Skipping line: {e}")

    return data
```

```python
file_product = "data/meta_Movies_and_TV.jsonl"
data = load_metadata(file_product)
df_meta = pd.DataFrame(data, columns=[
        "main_category", "title", "subtitle", "average_rating", "rating_number",
        "features", "description", "price", "images",
        "details", "parent_asin", "bought_together"
    ])
```

Filtering the datasets

```python
# reviews
review_length = 10
review_threshold = 15
helpful_threshold = 1
vertified_threshold = True

# moves
title_threshold = True # Needs to have a title
```

```python
# Remove rows where 'helpful_vote' is 0
df_cleaned = df_review[df_review['helpful_vote'] >= helpful_threshold]
print(f"Shape after removing unhelpful: {df_cleaned.shape}")

# Remove rows with less than `review_length` words in 'text'
df_cleaned = df_cleaned[df_cleaned['text'].apply(lambda x: len(str(x).split()) >= review_length)]
print(f"Shape after removing sub {review_length} word reviews: {df_cleaned.shape}")

# Remove rows where 'verified_purchase' is False
df_cleaned = df_cleaned[df_cleaned['verified_purchase'] == True]
print(f"Shape after removing unverified purchases: {df_cleaned.shape}")

# Calculate the total number of words in the 'text' column after cleaning
total_words = df_cleaned['text'].apply(lambda x: len(str(x).split())).sum()
print(f"Total number of words in the dataset after cleaning: {total_words:,}")

# Count reviews per ASIN
asin_review_counts = df_cleaned['parent_asin'].value_counts()

# Filter ASINs based on the review threshold
valid_asins = asin_review_counts[asin_review_counts >= review_threshold].index

# Keep only rows where the ASIN is in the list of valid ASINs
df_cleaned = df_cleaned[df_cleaned['parent_asin'].isin(valid_asins)]
print(f"Shape after filtering products with less than {review_threshold} reviews: {df_cleaned.shape}")

print(f"Unique ASINs after filtering: {df_cleaned['parent_asin'].nunique()}")
```

```python
df_merged = pd.merge(df_cleaned, df_meta_cleaned, on="parent_asin", how='inner')
df_merged.shape
```
